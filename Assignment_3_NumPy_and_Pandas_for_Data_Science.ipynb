```python
#Task 1.1: Data Loading and Exploration
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv('StudentsPerformance.csv')

# Your tasks:
# 1. Display first 10 rows
# 2. Show dataset shape and info()
# 3. Display unique values in each categorical column
# 4. Show basic statistics for all score columns

# Your code here:
#1. Display first 10 rows
print(df[:10])
print()
#2. Show dataset shape and info()
print(df.shape)
print()
print(df.info())
#3. Display unique values in each categorical column
columns = df.select_dtypes(include=['object']).columns
for i in columns:
    unique_values = df[i].unique()
    print(f"{i}: {unique_values}")
#4show basic statistics for all score columns
print()
print(df.describe())

```

```python
#Task 1.2: NumPy Array Operations
# Convert score columns to numpy arrays
mscores = df['math score'].values
rscores = df['reading score'].values
wscores = df['writing score'].values

# Your tasks:
# 1. Calculate mean, median, std, min, max for each subject using NumPy
# 2. Find total students who scored above 80 in math
# 3. Find total students who scored below 50 in any subject
# 4. Calculate the overall average score across all three subjects

# Expected output example:
# Math - Mean: 66.1, Median: 66.0, Std: 15.2
# Students with math > 80: 132
# Students with any score < 50: 178

# Your code here:
#1. Calculate mean, median, std, min, max for each subject using NumPy
subjects = {'Math': mscores, 'Reading': rscores, 'Writing': wscores}
for subject, scores in subjects.items():
    mean = np.mean(scores)
    median = np.median(scores)
    std = np.std(scores)
    min_score = np.min(scores)
    max_score = np.max(scores)
    print(f"{subject} - Mean: {mean:.1f}, Median: {median:.1f}, Std: {std:.1f}, Min: {min_score}, Max: {max_score}")
#2. Find total students who scored above 80 in math
m80plus = np.sum(mscores > 80)
print()
print(f"Students with math > 80: {m80plus}")
#3. Find total students who scored below 50 in any subject
below50 = np.sum((mscores < 50) | (rscores < 50) | (wscores < 50))
print()
print(f"Students with any score < 50: {below50}")
```

```python
# Task 2.1: Identify and Handle Missing Data
# Check the dataset for any missing values
# Your tasks:
# 1. Check for null values in each column using .isnull().sum()
# 2. Check for any impossible values (scores > 100 or < 0)
# 3. Display summary of data quality

# Note: This dataset is clean, but let's verify!

# Your code here:
#1. Check for null values in each column using .isnull().sum()
mscores = df['math score'].values
rscores = df['reading score'].values
wscores = df['writing score'].values
print(df.isnull().sum())
print()
#2. Check for any impossible values (scores > 100 or < 0)
invalid_math = np.sum((mscores < 0) | (mscores > 100))
invalid_reading = np.sum((rscores < 0) | (rscores > 100))
invalid_writing = np.sum((wscores < 0) | (wscores > 100))
print(f"Invalid Math Scores: {invalid_math}")
print(f"Invalid Reading Scores: {invalid_reading}")
print(f"Invalid Writing Scores: {invalid_writing}")
print()
#3. Display summary of data quality
print(df.info())
```

```python
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np
import pandas as pd

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

# Steps to follow:
# Step 1: Create copy and show original missing values
df = pd.read_csv('StudentsPerformance.csv')
df_practice = df.copy()
print("Original missing values:", df_practice.isnull().sum())
mscores = df['math score'].values
rscores = df['reading score'].values
wscores = df['writing score'].values

# Step 2: Create missing data
np.random.seed(42)  # For reproducible results
missing_math_idx = np.random.choice(df_practice.index, 50)
#step 3
missing_writing_idx = np.random.choice(df_practice.index, 30)
print(missing_math_idx)
print(missing_writing_idx)

# Continue the task...
# Your code here:
#4. Fill missing math scores with the median math score:
df_practice["math score"] = df_practice["math score"].replace(df_practice["math score"].median(), np.nan)
print(df_practice["math score"])
#5 Filling missing writing score with mean writing score:
df_practice["writing score"] = df_practice["writing score"].replace(df_practice["writing score"].mean(), np.nan)
print(df_practice["writing score"])
```

```python
#Task 3.2: Basic Groupby Analysis
# Comprehensive groupby analysis
# Your tasks:
import numpy as np
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

# Your code here:
#1. Group by 'gender' and calculate average, count of students
grouped = df.groupby('gender')[['math score', 'reading score', 'writing score']].agg(['mean', 'count'])
print("grouped by gender: \n", grouped)
print()
#2. Group by 'test preparation course' and show mean scores
pres_grouped = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print("grouped by test preparation course: \n", pres_grouped)
print()
#3. Group by 'parental level of education' and calculate average math score
par_grouped = df.groupby('parental level of education')['math score'].mean().sort_values()
print("grouped by parental level of education :\n", par_grouped)
print()#4. Group by 'lunch' type and show average scores, count of students
lunch_grouped = df.groupby('lunch')[['math score', 'reading score', 'writing score']].agg(['mean', 'count'])
print("grouped by lunch:\n", lunch_grouped)

```

```python
#Task 3.3: Use your data analysis skills to answer these easy questions:

# 1. Who performs better in math - males or females?
#    - Calculate average math score by gender
#    - Show the difference between male and female average

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
#    - Rank the subjects from highest to lowest average score

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
#    - Identify which education level leads to highest student performance
import numpy as np
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
# 1. Who performs better in math male or female?
gpgender = df.groupby('gender')['math score'].mean()
print("Average math score of :", gpgender)
print("difference between male math score and female math score average:", round((abs(gpgender['male'] - gpgender['female'])), 2))
print()
if gpgender['male'] > gpgender['female']:
    print("Males perform better in math")
else:
    print("Females perform better in math")
print()
# 2. Which subject do students perform best in overall?
overall_avg = df[['math score', 'reading score', 'writing score']].mean().sort_values()
print("Overall average scores:\n", overall_avg)
print("Students perform best in:", overall_avg.idxmax(), "-", overall_avg.max())
print()
print("students perform medium in:", overall_avg.index[1], "-", overall_avg.iloc[1])
print()
print("students perform worst in:", overall_avg.idxmin(), "-", overall_avg.min())
print()
# 3. What's the impact of parental education?
avgbypredu = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
# Calculate total score by summing the mean scores for each subject
avgbypredu['total score'] = avgbypredu[['math score', 'reading score', 'writing score']].sum(axis=1)
print("Average total score by parental education level:\n", avgbypredu['total score'])
highest_edu = avgbypredu['total score'].idxmax()
print("Highest student performance is from parental education level:  ", highest_edu, "-", avgbypredu['total score'].max())
print()
lowest_edu = avgbypredu['total score'].idxmin()
print("The lowest student performance is from parental education level:  ", lowest_edu, "-", avgbypredu['total score'].min())
print()
```
